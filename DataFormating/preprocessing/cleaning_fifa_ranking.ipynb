```python
import pandas as pd
```

```python
DIR = "../../datasets/"

fifa_ranking = pd.read_csv(DIR+"fifa_ranking.csv", index_col=0)
```

```python
fifa_ranking.head()
```

```python
fifa_ranking.drop(["country_abrv", "confederation"], axis=1, inplace=True)
fifa_ranking.reset_index(inplace=True)
```

```python
fifa_ranking["year"] = [int(x[:4]) for x in fifa_ranking["rank_date"]]
fifa_ranking.drop("rank_date", axis=1, inplace=True)
```

```python
fifa_ranking.columns = [
    "rank",
    "country",
    "total_points", 
    "previous_points",
    "rank_change",
    "cur_year_avg",
    "cur_year_avg_weighted",
    "last_year_avg",
    "last_year_avg_weighted",
    "two_year_ago_avg",
    "two_year_ago_weighted",
    "three_year_ago_avg",
    "three_year_ago_weighted",
    "year"
]
```

```python
fifa_ranking.head()
```

# Save into a CSV file

```python
fifa_ranking.to_csv("../fifa_ranking.csv", index=0)
```
