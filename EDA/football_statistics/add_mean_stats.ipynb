```python
import pandas as pd

from mean_stats import get_average_goals
```

# Import `compressed_final` dataset

```python
compressed_final = pd.read_csv("../compressed_final.csv")
```

```python
compressed_final.head()
```

# Add mean statistic to dataset

```python
new_compressed_final = pd.DataFrame(
    columns=list(compressed_final.columns) + ["Mean Home Team Goals", "Mean Away Team Goals"]
)
```

```python
for i in range(len(compressed_final)):
    data = list(compressed_final.iloc[i])
    res = get_average_goals(data[3], data[6])
    if res != 0:
        data.append(res[0])
        data.append(res[1])
    else:
        data.append(0)
        data.append(0)
        
    new_compressed_final = new_compressed_final.append(
        pd.Series(data, index=new_compressed_final.columns),
        ignore_index=True
    )
```

```python
compressed_final = new_compressed_final

compressed_final.head()
```

# Save datasets into CSV files

```python
compressed_final.to_csv("../compressed_final.csv", index=0)
```
