#### I'm not shure about   Stage, Assistant 1, Assistant 2

```python
wcm.drop(["Win conditions", "RoundID", "MatchID", "Stage", "Assistant 1",
          "Assistant 2", "Home Team Initials", "Away Team Initials"], axis=1, inplace=True)
```

```python
new_wcm = pd.DataFrame(columns=wcm.columns)
for i in wcm.index:
    if int(wcm.iloc[i]["Year"]) >= 2010:
        new_wcm = new_wcm.append(wcm.iloc[i])
```

```python
wcm = new_wcm
len(wcm)
```

```python
wcm = wcm.reset_index()
wcm.drop(["index", "level_0"], axis=1, inplace=True)
```
